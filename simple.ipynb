```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Activation
%matplotlib inline

```

```python
x1 = np.linspace(10,-5,30)
#x1 = np.array([np.arange(1,6,1)])
#y1 = np.array([np.arange(100,600,100)])
y1 = np.linspace(400,2000,30)
print(x1)
print(y1)
#print(preprocessing.normalize(y1))
#x2 = preprocessing.normalize(x1)
#y2 = preprocessing.normalize(y1)
#print(x2[0])
```

```python
model = Sequential()
model.add(Dense(1, input_dim=1,  activation='linear'))
model.compile(optimizer='sgd',loss='mse')
model.fit(x1,y1, epochs=600, verbose=1)
```

```python
weights = model.layers[0].get_weights()
w = weights[0][0][0]
b = weights[1][0]

plt.scatter(x1,y1)
plt.grid(True)
plt.ylabel('Casacos vendidos')
plt.xlabel('temperatura')
plt.title('Temperatura X Casacos vendidos')
plt.plot(x1, w*x1 + b, c='red')
plt.show()
```

```python
print(model.predict(np.array([-8])))
```
